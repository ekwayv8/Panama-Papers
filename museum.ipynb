# 2) Voici les réponses à tes fameuses questions

### a) Combien y-a-t-il de musées en France métropolitaine ?

````python
#Commençon déjà par visualiser notre table
df_museum
````

````python
# Essayons de répondre à ta question en effectuant un décompte 
number_of_museum = df_museum['osm_id'].value_counts(dropna=False).size
#Et n'oublions pas d'afficher le résultat avec un bon view print des familles
result = str(number_of_museum)

print ("Answer => il y a " + result + " musées en France métropolitaine")
````

### b) Dans quelle(s) ville(s) y-a-t-il le plus de musées ?

````python
# Alors cher manager nous allons commencer par compter le nombre de musées par ville

number_per_city = df_museum['city'].value_counts()
highest_value =str(max(number_per_city))
print ("Dans une seule ville nous retrouvons jusqu'à " + highest_value + " Musées...")
````

````python
hash_number_per_city = number_per_city.to_dict()
````

````python
hash_number_per_city
````

````python
city_with_highest_value = list(hash_number_per_city)[0]
print ("... et la ville en question est", city_with_highest_value)
````

### c) Quel est le nombre moyen de musées par ville ?

````python
# Nous allons maintenant effectuer 2 , 3 stats de base

mean_per_city = number_per_city.mean()
print ("En moyenne nous retrouvons " + str(mean_per_city) + " musées par ville")
````

### d) Quel est le nombre médian de musées par ville ?

````python
median_per_city = number_per_city.median()
print ("La médiane correspond à " + str(median_per_city) + " musée par ville")
````

### e) Comment sont répartis les musées par type (en pourcentage) ?

````python
#ci-dessous la répartition attendue
repartition = df_museum['tags'].value_counts(normalize = 'true')*100

print ("Voici la réparition en pourcentage % des différentes catégories ou types de musées" ) 

repartition
````

### f) Combien y-a-t-il de musées dont le nom commence par "Château" ?

````python
museum_name = df_museum['name'].str.startswith('Château').value_counts()
 
print(museum_name)

print (" ")

print("Nous avons donc 57 musées dont le nom commence par Château" )
````

### g) Pour combien de musées dispose-t-on de l'adresse du site web ?

````python
museum_with_website = df_museum['website'].value_counts().sum()

print("nous disposons de l'adresse du site web de " + str(museum_with_website) + " musées")
````

### h) Quel département français possède le plus de musées sur son territoire ?


````python
department_hash = df_museum['postal_code'].astype(str).str[0:2].value_counts().to_dict()

department_with_most_museum = list(department_hash)[0]
print("Le département français qui possède le plus de musées sur son territoire est le " + str(department_with_most_museum))
````

### i) Quel département français possède le moins de musées sur son territoire ?

````python
department_with_less_museum = list(department_hash)[-1]
print("Le département français qui possède le moins de musées sur son territoire est le " + department_with_less_museum)
                                   
````
